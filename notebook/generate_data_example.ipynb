```python
import numpy as np
import pandas as pd
from itertools import permutations, combinations
```

```python
Sentences = {
     "4":["As hobbies, John likes {}, {}, {}, and {} in his free time.",
          "John's free-time hobbies are {}, {}, {}, and {}.",
          "In his free time, John likes {}, {}, {}, and {} as hobbies."],
     "5":["As hobbies, John likes {}, {}, {}, {}, and {} in his free time.",
          "John's free-time hobbies are {}, {}, {}, {}, and {}.",
          "In his free time, John likes {}, {}, {}, {}, and {} as hobbies."],
     "6":["As hobbies, John likes {}, {}, {}, {}, {}, and {} in his free time.",
          "John's free-time hobbies are {}, {}, {}, {}, {}, and {}.",
          "In his free time, John likes {}, {}, {}, {}, {}, and {} as hobbies."]
}
Hobbies = {"1" :["running", "jogging"],
    "2" :["Drone flying", "piloting Aerial drones"],
    "3" :["jazzercise", "aerobics"],
    "4" :["making pottery", "making ceramics"],
    "5" :["water gardening", "aquatic gardening"],
    "6" :["caving", "spelunking", "potholing"],
    "7" :["cycling", "bicycling", "biking"],
    "8" :["reading"],
    "9" :["writing journals", "journal writings", "journaling"],
    "10":["sculling", "rowing"]
 }
```

```python
def make_choices(Content,size):
    choices_set = []
    for i in size:
        choices = list(combinations(Content.keys(),i))
        np.random.shuffle(choices)
        make_choices = "_".join(list(choices[-1]))
        while make_choices in choices_set:
            choices.drop(-1)
            make_choices = "_".join(list(choices[-1]))
        choices_set.append(make_choices)
    return choices_set

def generate_a_sentence(Sentences, Content, current_content_choice):
    size = str(len(current_content_choice))
    np.random.shuffle(current_content_choice)
    generating_sentence = np.random.choice(Sentences[size],1)[0]
    contents = []
    for i in current_content_choice:
        contents.append(np.random.choice(Content[i],1)[0])
    sentence = generating_sentence.format(*contents)
    return sentence
```

```python

def build_data(N, p, Sentences, Content):
    texts = []
    label = []
    K = len(p)
    N_k = np.random.multinomial(N,p)
    size = [int(x) for x in np.random.choice(list(Sentences.keys()),K)]
    choices_set = make_choices(Content, size)
    for n_k, cs in zip(N_k,choices_set):
        current_content_choice = cs.split("_")
        for i in range(n_k):
            text = generate_a_sentence(Sentences, Content, current_content_choice)
            texts.append(text)
            label.append(cs)
    return pd.DataFrame.from_dict({"text":texts, "labels":label})
def build_comparison(text_data):
    text_1 = []
    text_2 = []
    label_1 = []
    label_2 = []
    idx_1   = []
    idx_2   = []
    for i in combinations(range(text_data.shape[0]), 2):
        idx_1.append(i[0])
        idx_2.append(i[1])
        text_1.append(text_data["text"].iloc[i[0]])
        text_2.append(text_data["text"].iloc[i[1]])
        label_1.append(text_data["labels"].iloc[i[0]])
        label_2.append(text_data["labels"].iloc[i[1]])
    return pd.DataFrame.from_dict({"index_1":idx_1, "index_2":idx_2,"text_1":text_1, "text_2":text_2, "label_1":label_1,"label_2":label_2 })
```

```python
np.random.seed(1)
text_data =build_data(30,[0.5,0.5],Sentences, Hobbies)
```

```python
build_comparison(text_data)
```
